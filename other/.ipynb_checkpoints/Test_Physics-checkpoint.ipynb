```python
import matplotlib.pyplot as plt
import numpy as np 
```

# Moteur 2D

Lancer : 

`make && ./Moteur2D`


```python
v = 20

def f(t) :
    return min(1.5*v*t+1, v)/v

def g(t):
    return max(0.3*v*t, 0)/v


N = 15
ts = range(N)


decc = np.zeros(N)
decc[0] = 1.
for i in range(N-1):
    decc[i+1] = g(decc[i])

    
acc = np.zeros(N)
for i in range(N-1):
    acc[i+1] = f(acc[i])

plt.figure(figsize=(15,5))
plt.subplot(121)
plt.plot(ts,v*acc,label="v(t)")
plt.plot(ts,[int(v*ac) for ac in acc], label="v(t) entier")

plt.legend()

plt.subplot(122)
plt.plot(ts,v*decc,label="v(t)")
plt.plot(ts,[int(v*dec) for dec in decc], label="v(t) entier")
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np 
```

```python
def f(t):
    return min(t*2+0.02 , 1)
def g(t):
    #return max(1 - ((1-t)*2+0.02) , 0)
    return max(t*0.25-0.005 , 0)
N = 10
ts = range(N)


decc = np.zeros(N)
decc[0] = 1.
for i in range(N-1):
    decc[i+1] = g(decc[i])

acc = np.zeros(N)
for i in range(N-1):
    acc[i+1] = f(acc[i])
    
plt.plot(ts,acc)
plt.plot(ts,decc)
plt.show()

```
